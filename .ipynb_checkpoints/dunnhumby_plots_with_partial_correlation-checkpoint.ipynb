```python
import networkx as nx
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
mat = scipy.io.loadmat('precision_from_cov_24_dunnhumby_64_500_0000005.mat')

#precision-18-votes-4v2.mat --> good result , 100 epochs,4 mix comp, 
# precision-18-votes-4v3.mat --> first one is good 
```

```python
adj = mat['saveA']
```

```python
pc = np.zeros((24,24))
adj0 =adj[0]
for i in range(24):
    for j in range(24):
        if i!=j:
            pc[i,j] = -adj0[i,j]/np.sqrt(adj0[i,i]*adj0[j,j]) 
            
           # pc[i,j] = -adj0[i,j]/np.sqrt(adj0[i,i],adj0[j,j])
```
