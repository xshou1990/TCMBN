```python
import networkx as nx
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd
```

```python
# load adjacency matrix
mat = scipy.io.loadmat('precision_from_cov_24_dunnhumby_64_0_0000005.mat')
adj = mat['saveA']
```

```python
# compute partial correlation
pc = np.zeros((24,24))
adj0 =adj[0]
for i in range(24):
    for j in range(24):
        if i!=j:
            pc[i,j] = -adj0[i,j]/np.sqrt(adj0[i,i]*adj0[j,j])          
```

```python
## all 24 department
info_party = np.array(['UNKNOWN', 'CHEF SHOPPE', 'COSMETICS', 'COUP/STR & MFG', 'DELI', 'DRUG GM', 'FLORAL', 'FROZEN GROCERY', 'GARDEN CENTER', 'GROCERY', 'KIOSK-GAS', 'MEAT', 'MEAT-PCKGD', 'MISC SALES TRAN', 'MISC. TRANS.', 'NUTRITION', 'PASTRY', 'PRODUCE', 'RESTAURANT', 'SALAD BAR', 'SEAFOOD', 'SEAFOOD-PCKGD', 'SPIRITS', 'TRAVEL & LEISUR'])
```

```python
# take a subset
subset = [ 'COSMETICS', 'DELI', 'DRUG GM', 'FLORAL', 'FROZEN GROCERY', 'GARDEN CENTER',  'KIOSK-GAS', 'MEAT',   'NUTRITION', 'PASTRY', 'PRODUCE', 'RESTAURANT', 'SALAD BAR', 'SEAFOOD', 'SPIRITS', 'TRAVEL & LEISUR'][5:13]
subset = np.sort([info_party.tolist().index(x) for x in subset])
# subset of partial correlation
new_pc = (pc[subset][:,subset]).round(3)
```

```python
# plot the network

plt.figure(figsize=(10,5))

G = new_pc 
A = nx.from_numpy_matrix(G)


pos = nx.spring_layout(nx.Graph(A), k=5/math.sqrt(A.order()), iterations=100)

labels = info_party[subset].tolist()
lab_dict = dict(zip( list(range(0,len(subset))), labels))
nx.draw_networkx(A,pos,labels = lab_dict,font_size = 14)

# Get weights of each edge and assign to labels
edge_labels = nx.get_edge_attributes(A, "weight")

# Draw edge labels using layout and list of labels
nx.draw_networkx_edge_labels(G, pos=pos, edge_labels=edge_labels,font_size = 14,label_pos=0.4)

plt.margins(x=0.11)
```
